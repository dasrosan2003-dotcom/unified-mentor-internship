```python
import csv

# ==========================================
# Load Resumes from CSV
# ==========================================

def load_resumes(filename):

    candidates = []

    try:

        with open(filename, "r") as file:

            reader = csv.DictReader(file)

            for row in reader:

                candidate = {
                    "name": row["Name"],
                    "skills": [skill.strip().lower()
                               for skill in row["Skills"].split(",")]
                }

                candidates.append(candidate)

    except FileNotFoundError:
        print("Resume file not found.")

    return candidates


# ==========================================
# Calculate Match Percentage
# ==========================================

def calculate_match(candidate_skills, required_skills):

    matched_skills = []

    for skill in required_skills:

        if skill.lower() in candidate_skills:
            matched_skills.append(skill)

    match_percentage = (len(matched_skills)
                        / len(required_skills)) * 100

    return round(match_percentage, 2), matched_skills


# ==========================================
# Shortlist Candidates
# ==========================================

def shortlist_candidates(candidates, required_skills):

    shortlisted = []
    rejected = []

    for candidate in candidates:

        match_percentage, matched_skills = calculate_match(
            candidate["skills"],
            required_skills
        )

        candidate_data = {
            "name": candidate["name"],
            "match": match_percentage,
            "matched_skills": matched_skills
        }

        # Selection Criteria
        if match_percentage >= 60:
            shortlisted.append(candidate_data)
        else:
            rejected.append(candidate_data)

    # Rank by match percentage
    shortlisted.sort(key=lambda x: x["match"], reverse=True)

    return shortlisted, rejected


# ==========================================
# Export Shortlisted Candidates
# ==========================================

def export_to_csv(shortlisted):

    with open("shortlisted_candidates.csv",
              "w",
              newline="") as file:

        writer = csv.writer(file)

        writer.writerow([
            "Name",
            "Match Percentage",
            "Matched Skills"
        ])

        for candidate in shortlisted:

            writer.writerow([
                candidate["name"],
                f"{candidate['match']}%",
                ", ".join(candidate["matched_skills"])
            ])

    print("\nShortlisted candidates exported successfully.")


# ==========================================
# Display Candidates
# ==========================================

def display_candidates(title, candidates):

    print("\n" + "=" * 50)
    print(title)
    print("=" * 50)

    if not candidates:
        print("No candidates found.")
        return

    for index, candidate in enumerate(candidates, start=1):

        print(f"\n{index}. {candidate['name']}")
        print(f"Match Percentage : {candidate['match']}%")
        print("Matched Skills   :",
              ", ".join(candidate["matched_skills"]))


# ==========================================
# Main Program
# ==========================================

def main():

    print("=" * 55)
    print("      AI RESUME SHORTLISTING SYSTEM")
    print("=" * 55)

    # Load resumes
    candidates = load_resumes(r"C:\Users\dasro\unified mentor internship\resumes.csv")

    if not candidates:
        return

    # User input skills
    skills_input = input(
        "\nEnter required skills (comma separated): "
    )

    required_skills = [
        skill.strip().lower()
        for skill in skills_input.split(",")
    ]

    # Shortlist
    shortlisted, rejected = shortlist_candidates(
        candidates,
        required_skills
    )

    # Display results
    display_candidates(
        "SELECTED CANDIDATES",
        shortlisted
    )

    display_candidates(
        "NOT SELECTED CANDIDATES",
        rejected
    )

    # Export option
    export = input(
        "\nDo you want to export shortlisted candidates? (yes/no): "
    )

    if export.lower() == "yes":
        export_to_csv(shortlisted)

    print("\nProcess Completed.")


# ==========================================
# Run Program
# ==========================================

if __name__ == "__main__":
    main()
```
