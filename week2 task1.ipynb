```python
import datetime
import random
import os

# -----------------------------
# Command History List
# -----------------------------
command_history = []

# -----------------------------
# Save Command to File
# -----------------------------
def save_history(command):
    with open("history.txt", "a") as file:
        file.write(command + "\n")


# -----------------------------
# Show Current Time
# -----------------------------
def show_time():
    current_time = datetime.datetime.now().strftime("%H:%M:%S")
    print(f"\nCurrent Time: {current_time}")


# -----------------------------
# Show Current Date
# -----------------------------
def show_date():
    current_date = datetime.datetime.now().strftime("%d-%m-%Y")
    print(f"\nCurrent Date: {current_date}")


# -----------------------------
# Help Menu
# -----------------------------
def show_help():
    print("""
Available Commands:
-------------------
time        -> Show current time
date        -> Show current date
help        -> Show command list
calculator  -> Open calculator
joke        -> Tell a joke
add note    -> Save a note
view notes  -> View saved notes
history     -> Show command history
exit        -> Exit program
""")


# -----------------------------
# Calculator
# -----------------------------
def calculator():
    try:
        num1 = float(input("Enter first number: "))
        operator = input("Enter operator (+, -, *, /): ")
        num2 = float(input("Enter second number: "))

        if operator == "+":
            print("Result:", num1 + num2)

        elif operator == "-":
            print("Result:", num1 - num2)

        elif operator == "*":
            print("Result:", num1 * num2)

        elif operator == "/":
            if num2 != 0:
                print("Result:", num1 / num2)
            else:
                print("Cannot divide by zero")

        else:
            print("Invalid operator")

    except ValueError:
        print("Invalid input")


# -----------------------------
# Random Joke
# -----------------------------
def tell_joke():
    jokes = [
        "Why do programmers prefer dark mode? Because light attracts bugs!",
        "Why was the computer cold? It forgot to close Windows.",
        "Why do Java developers wear glasses? Because they don't C#."
    ]

    print("\n", random.choice(jokes))


# -----------------------------
# Add Note
# -----------------------------
def add_note():
    note = input("Write your note: ")

    with open("notes.txt", "a") as file:
        file.write(note + "\n")

    print("Note saved successfully.")


# -----------------------------
# View Notes
# -----------------------------
def view_notes():
    if os.path.exists("notes.txt"):
        with open("notes.txt", "r") as file:
            notes = file.readlines()

            if notes:
                print("\nSaved Notes:")
                for index, note in enumerate(notes, start=1):
                    print(f"{index}. {note.strip()}")

            else:
                print("No notes found.")

    else:
        print("No notes file found.")


# -----------------------------
# Show Command History
# -----------------------------
def show_history():
    if command_history:
        print("\nCommand History:")
        for index, cmd in enumerate(command_history, start=1):
            print(f"{index}. {cmd}")
    else:
        print("No commands used yet.")


# -----------------------------
# Exit Program
# -----------------------------
def exit_program():
    print("\nClosing Smart CLI Assistant...")
    exit()


# -----------------------------
# Dictionary-Based Command Handling
# -----------------------------
commands = {
    "time": show_time,
    "date": show_date,
    "help": show_help,
    "calculator": calculator,
    "joke": tell_joke,
    "add note": add_note,
    "view notes": view_notes,
    "history": show_history,
    "exit": exit_program
}


# -----------------------------
# Natural Language Matching
# -----------------------------
def match_command(user_input):

    user_input = user_input.lower()

    if "time" in user_input:
        return "time"

    elif "date" in user_input:
        return "date"

    elif "help" in user_input:
        return "help"

    elif "calculate" in user_input or "calculator" in user_input:
        return "calculator"

    elif "joke" in user_input:
        return "joke"

    elif "add note" in user_input:
        return "add note"

    elif "view note" in user_input or "notes" in user_input:
        return "view notes"

    elif "history" in user_input:
        return "history"

    elif "exit" in user_input or "quit" in user_input:
        return "exit"

    else:
        return None


# -----------------------------
# Main Program
# -----------------------------
def main():

    print("=" * 40)
    print("   SMART CLI ASSISTANT")
    print("=" * 40)
    print("Type 'help' to see commands.\n")

    while True:

        user_input = input("You: ").strip().lower()

        command_history.append(user_input)
        save_history(user_input)

        matched_command = match_command(user_input)

        if matched_command in commands:
            commands[matched_command]()

        else:
            print("Unknown command. Type 'help'.")


# -----------------------------
# Run Program
# -----------------------------
if __name__ == "__main__":
    main()
```
