```python
import requests
import os

# =========================================
# OpenWeather API Key
# =========================================

API_KEY = "4a80be0ed17d776ae178575c734a8fd1"

# Example:
# API_KEY = "4a80be0ed17d776ae178575c734a8fd1"

BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

FORECAST_URL = "https://api.openweathermap.org/data/2.5/forecast"


# =========================================
# Save searched cities
# =========================================

def save_city(city):

    with open("searched_cities.txt", "a") as file:
        file.write(city + "\n")


# =========================================
# Current Weather Function
# =========================================

def get_weather(city):

    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }

    try:

        response = requests.get(BASE_URL, params=params)

        data = response.json()

        # Success
        if response.status_code == 200:

            temperature = data["main"]["temp"]
            humidity = data["main"]["humidity"]
            condition = data["weather"][0]["description"]

            print("\n================================")
            print(f" Weather Report for {city}")
            print("================================")

            print(f"Temperature : {temperature}°C")
            print(f"Condition   : {condition}")
            print(f"Humidity    : {humidity}%")

            save_city(city)

        else:
            print("\nError:", data.get("message"))

    except requests.exceptions.RequestException:
        print("\nNetwork Error. Check internet connection.")


# =========================================
# 5-Day Forecast Function
# =========================================

def get_forecast(city):

    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }

    try:

        response = requests.get(FORECAST_URL, params=params)

        data = response.json()

        if response.status_code == 200:

            print("\n================================")
            print(f" 5-Day Forecast for {city}")
            print("================================")

            forecasts = data["list"]

            # Every 8th entry = 24 hours
            for item in forecasts[::8]:

                date = item["dt_txt"]
                temp = item["main"]["temp"]
                condition = item["weather"][0]["description"]

                print(f"\nDate        : {date}")
                print(f"Temperature : {temp}°C")
                print(f"Condition   : {condition}")

        else:
            print("\nError:", data.get("message"))

    except requests.exceptions.RequestException:
        print("\nNetwork Error.")


# =========================================
# Show Search History
# =========================================

def show_history():

    if os.path.exists("searched_cities.txt"):

        with open("searched_cities.txt", "r") as file:

            cities = file.readlines()

            if cities:

                print("\nSearch History")
                print("================")

                for index, city in enumerate(cities, start=1):
                    print(f"{index}. {city.strip()}")

            else:
                print("\nNo history found.")

    else:
        print("\nNo history file found.")


# =========================================
# Main Program
# =========================================

def main():

    print("=" * 40)
    print("        LIVE WEATHER APP")
    print("=" * 40)

    while True:

        print("""
1. Current Weather
2. 5-Day Forecast
3. Search History
4. Exit
""")

        choice = input("Enter choice: ")

        # Current Weather
        if choice == "1":

            city = input("Enter city name: ")
            get_weather(city)

        # Forecast
        elif choice == "2":

            city = input("Enter city name: ")
            get_forecast(city)

        # History
        elif choice == "3":

            show_history()

        # Exit
        elif choice == "4":

            print("\nExiting Weather App...")
            break

        else:
            print("\nInvalid choice. Try again.")


# =========================================
# Run Program
# =========================================

if __name__ == "__main__":
    main()
```
